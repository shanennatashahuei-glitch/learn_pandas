#13: Konversi nilai numerik ke dalam kategori pada data frame

```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

Pengelompokkan nilai numerik ke dalam beberapa kategori menggunakan cut():

```python
df['kelompok_usia'] = pd.cut(df['usia'], 
                             bins=[0, 18, 65, 99], 
                             labels=['anak', 'dewasa', 'manula'])
df
```

#14: Menggabungkan dua Pandas data frame

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1, 4])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0, 3])
df2
```

Menggabungkan dua data frame:

```python
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
df_outer = pd.merge(df1, df2, how='outer')
df_outer
```

#15: Memecah nilai string suatu kolom ke dalam beberapa kolom baru pada Pandas data frame

```python
data = {'nama':['Shanen Natasha', 'Wallace Quincy', 'Mbah Surip'], 
        'tempat_kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojokerto, Jawa Timur']}
df = pd.DataFrame(data)
df
```

Memecah nama depan dan nama belakang:

```python
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

Memecah nama kota dan provinsi:

```python
df[['kota', 'propinsi']] = df['tempat_kelahiran'].str.split(',', expand=True)
df
```

#16: Menata ulang data frame dengan multiple indexes menggunakan unstack

```python
df = pd.read_csv('titanicfull.csv')
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

Data frame dengan multiple indexes dari hasil groupping:

```python
df.groupby(['sex', 'pclass'])['survival_status'].mean()
```

Menata ulang data frame dengan mutiple indexes:

```python
df.groupby(['sex', 'pclass'])['survival_status'].mean().unstack()
```
