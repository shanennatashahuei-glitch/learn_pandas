#37: Seleksi baris dengan banyak kriteria pada Pandas data frame

```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.read_csv('titanicfull.csv')
df.head()
```

```python
df[
    (df['Sex'] == 'female') &
    (df['Age'] >= 60) &
    (df['Embarked'] == 'S') &
    (df['Survival status'] == 1)
]
```

```python
kr1 = df['Sex'] == 'female'
kr2 = df['Age'] >= 60
kr3 = df['Embarked'] == 'S'
kr4 = df['Survival status'] == 1

df[kr1 & kr2 & kr3 & kr4]
```

#39: Mengacak urutan baris pada Data Frame

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 6
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df.sample(frac=1.0, random_state=1)
```

```python
df.sample(frac=1.0, random_state=1).reset_index(drop=True) 
```

#40: Akses sekelompok data pada Data Frame yang sudah terkelompok dengan get_group()

```python
df = pd.read_csv('titanicfull.csv')
df.head()
```

```python
grouped_df = df.groupby('Sex')
```

```python
grouped_df.get_group('female').head(10)
```

```python
grouped_df = df.groupby('Survival status')
```

```python
grouped_df.get_group(1).head(10)
```
