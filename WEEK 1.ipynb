#02: Pemilihan baris (rows selection) pada data frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size = (n_rows, n_cols)), columns = cols)
df
```

Untuk membuat kolom A hanya terdiri dari angka 1 atau 3

```python
df[(df['A'] == 1)|(df['A'] == 3)]
```

Selection dengan fungsi isin(), sama seperti yang atas cuman berbeda fungsi

```python
df[df['A'].isin([1, 3])]
```

Bagaima untuk menampilkan baris yang tidak sama dengan 1 atau tidak sama dengan 3?

Operator negasi ~

```python
df[~df['A'].isin([1, 3])]
```

PANDAS 03

#03 Konversi tipe data String ke Numerik pada kolom data frame

```python
data = {'col1': ['1', '2', '3', 'teks'],
        'col2': ['1', '2', '3', '4']}
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

Konversi tipe data dengan fungsi astype()

```python
df_x = df.astype({'col2': 'float'})
df_x
```

```python
df_x.dtypes
```

Konversi tipe data numnerik dengan fungsi to_numeric(), coerce utk data yang tidak dapat diubah menjadi numerik

```python
df.apply(pd.to_numeric, errors = 'coerce')
```

PANDAS 04

#4 Pemilihan kolom (columns selection) pada data frame berdasarkan tipe data

```python
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size = (n_rows, n_cols)), columns = cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float')

dt = pd.date_range('2023-05-10', periods=(n_rows), freq='H')
df.index = pd.to_datetime(dt, format='%d%b%Y:%H:%M:%S.%f')

df['teks'] = list('ABCDE')
df
```

```python
df.dtypes
```

Cara untuk memilih kolom bertipe data numerik saja

```python
df.select_dtypes(include = 'number')
```

```python
df.select_dtypes(include = 'float')
```

```python
df.select_dtypes(include = 'int')
```

Cara untuk memilih kolom bertipe data String atau object

```python
df.select_dtypes(include='object')
```

Cara untuk memilih kolom bertipe data datetime

```python
df.select_dtypes(include='datetime')
```

Cara untuk menampilkan kolom dengan kombinasi tipe data

```python
df.select_dtypes(include= ['datetime', 'object'])
```
