#7: Menghapus missing values pada data frame

```python
import pandas as pd

print(pd.__version__)
```

```python
n_rows = 30
n_cols = 4
cols = ['A', 'B', 'C', 'D']

df = pd.DataFrame(np.random.randn(n_rows, n_cols), columns=cols)

nan_mask = np.random.rand(n_rows, n_cols) < 0.2
df[nan_mask] = np.nan

df = df.reset_index()
df.head()
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)
```

Menghapus setiap kolom yang mengandung missing values

```python
df = df.dropna(axis='columns') 
df.head()
```

Menghapus setiap baris yang mengandung missing values

```python
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

Persentase missing values untuk tiap kolom

```python
df = df_backup.copy(deep=True)
df.isna().mean()
```

Menghapus setiap kolom yang mengandung missing values berdasarkan threshold

```python
treshold = len(df) * 0.9
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```
